## The Quadratic Formula

Any second-order polynomial $p(x) = ax^2 + bx + c$ can be rewritten as a scaled, translated parabola

$p(x) = a(x + \frac{b}{2a}) ^2 + (c - \frac{b^2}{4a})$.

In this form, the roots can be found by solving $p(x) = 0$ to get

$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$.

This is the general _quadratic formula_.

```python
def find_quadratic_roots(a, b, c):
    return [(-b + sign * (b ** 2 - 4 * a * c) ** 0.5) / (2 * a) for sign in [-1, 1]]
```

```python
a = 1; b = -5; c = 6;
find_quadratic_roots(a, b, c)
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x_offset = -b / (2 * a)
x = np.linspace(x_offset - 1, x_offset + 1, 100)
y = a * x ** 2 + b * x + c
_ = plt.plot(x, y)
```

## Complex Roots

Let $a = 1$, $b = 0$, and $c = 4$.

Then $p(x) = x^2 + 4$.

This is a parabola centered at $x = 0$, and reaching upward to positive infinity, never going below 4. It has no real zeros:

```python
a = 1; b = 0; c = 4;

x_offset = -b / (2 * a)
x = np.linspace(x_offset - 1, x_offset + 1, 100)
y = a * x ** 2 + b * x + c
plt.title('An example parabola defined by $p(x) = x^2 + 4$')
_ = plt.plot(x, y)
```

But the quadratic formula says that the "roots" are given by $x = \pm \sqrt{-4} = \pm 2 \sqrt{-1}$.

If we define $j \triangleq \sqrt{-1}$,

then we can express the polynomial in terms of its roots as

$p(x) = (x + 2j)(x - 2j)$.

```python
imag_roots = find_quadratic_roots(a, b, c) # Python handles this well by default
imag_roots
```

```python
np.pi + 1j
```

**3. Convert the following complex numbers to _Cartesian form_, $x + jy$**

a) $(1 + j)^2$

$(1 + j)^2 = 1 + 2j + j^2 = 1 + 2j - 1 = 2j$

b) $\frac{1+j}{1-j}$

$\frac{1+j}{1-j} = \frac{(1+j)(1+j)}{(1-j)(1+j)} = \frac{1 + 2j + j^2}{1 - j^2} =  \frac{2j}{2} = j$

c) $\frac{a + jb}{c+jd}$

$\frac{a + jb}{c+jd} = \frac{(a + jb)(c - jd)}{(c+jd)(c - jd)} = \frac{ac - jad + jbc - j^2bd}{c^2 - j^2d^2} = \frac{ac - jad + jbc + bd}{c^2 + d^2} = j\frac{bc - ad}{c^2 + d^2} + \frac{ac + bd}{c^2 + d^2}$

d) $e^{j\pi/6}$

$e^{j\pi/6} = \cos{\pi/6} + j\sin{\pi/6} = \frac{\sqrt{3}}{2} + \frac{1}{2}j$

e) $5e^{j\pi/2}$

$5e^{j\pi/2} = 5\cos{\pi/2} + 5j\sin{\pi/2} = 5j$

f) $\frac{\bar{j^2}}{\bar{(1+j)^2}}$

$\frac{\bar{j^2}}{\bar{(1+j)^2}} = \frac{\bar{-1}}{\bar{2j}} = \frac{-1}{-2j} = \frac{1}{2}j$

g) $\ln(j)$

$\ln(j) = \ln((-1)^{1/2}) = \frac{1}{2} \ln(-1)$

Recall from question _(b)_ that $\frac{1+j}{1-j} = j$. Thus,

$\tan(\frac{1+j}{1-j}) = \tan(j) = \frac{\sin j}{\cos j}$.

Using the identities

$\cos(\theta) = \frac{e^{j\theta} + e^{-j\theta}}{2}$ and  
$\sin(\theta) = \frac{e^{j\theta} - e^{-j\theta}}{2j}$,

$\begin{align}\frac{\sin j}{\cos j} &= \frac{\frac{e^{j^2} - e^{-j^2}}{2j}}{\frac{e^{j^2} + e^{-j^2}}{2}} = \frac{\frac{e^{-1} - e}{2j}}{\frac{e^{-1} + e}{2}}\\

&= \frac{j(e^{-1} - e)}{j^2(e^{-1} + e)} = \frac{j(e^{-1} - e)}{-(e^{-1} + e)} \\
&= -j\frac{e^{-1} - e}{e^{-1} + e} = -j\frac{1 - e^2}{1 + e^2}
\end{align}$
